This notebook was used to determine the interquartile ranges for earthquake depth in the data set. These ranges were used to inform the scale for the visualiztion.


```python
import pandas as pd
import requests
from pprint import pprint
```

```python
earthquake_url = "https://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/4.5_month.geojson"
```

```python
response = requests.get(earthquake_url).json()
```

```python
earthquakes_df = pd.json_normalize(response['features'])
```

```python
earthquakes_df
```

```python
 coordinates_df = pd.DataFrame(earthquakes_df['geometry.coordinates'].to_list(),columns=['latitude','longitude','depth'])
```
